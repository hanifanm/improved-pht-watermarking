```python
def test_translation(min_p, max_p):
    
    px = np.random.uniform(low=min_p, high=max_p) * (1 if np.random.uniform() < 0.5 else -1)
    py = np.random.uniform(low=min_p, high=max_p) * (1 if np.random.uniform() < 0.5 else -1)
    
    (h, w, c) = my_image.image.shape
    dx = int(px * w)
    dy = int(py * h)
    
    img_translate = Attacker.translate_by_pixel(my_image.image, dx, dy)
    
    (cx, cy, cs) = AntiTranslation.calculate_center(my_image.image, img_translate)
    
    error_x = np.abs(dx + cx)
    error_y = np.abs(dy + cy)
    
    return (error_x, error_y)
```

```python
compilation_x = []
compilation_y = []
for min_p in np.arange(start = 0, stop = 0.5, step = 0.05):
    res_x = []
    res_y = []
    for i in range(5):
        (error_x, error_y) = test_translation(min_p, min_p + 0.05)
        res_x.append(error_x)
        res_y.append(error_y)
    compilation_x.append(res_x)
    compilation_y.append(res_y)
```

```python
result_x = []
for arr in compilation_x:
    mean = np.mean(arr)
    result_x.append(mean)
    
result_y = []
for arr in compilation_y:
    mean = np.mean(arr)
    result_y.append(mean)
```
